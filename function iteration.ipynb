```python
from pylab import plot,show
import numpy as np

def fixedpoint(f,x0,tol=10e-5,maxiter=100):
 
 e = 1
 itr = 0
 xp = []
 while(abs(e) > tol and itr < maxiter):
  x = f(x0)      
  e = np.linalg.norm(x0-x) # 差分
  x0 = x
  xp.append(x0)  # 追加
  itr = itr + 1
 return x,xp,itr
```

```python
def f(x):
    return np.sin(x)+x

x_start = 0.4
xf,xp,itr = fixedpoint(f,x_start)

x = np.linspace(0,4,100)
y = f(x)
plot(figsize=(14,10))
plot(x,y,xp,f(xp),'bo',
     x_start,f(x_start),'ro',xf,f(xf),'go',x,x,'k')
show()
```
